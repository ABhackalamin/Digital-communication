```python
# library 
import numpy as np
import matplotlib.pyplot as plot

# Number of bits
N = 8

# Generate random binary sequence
n = np.random.randint(2, size=N)
print('The binary sequence is', n)

# Signal Shaping
i = 1  # Bit index
a = 0  # Time interval start
t = np.arange(0, N + 0.01, 0.01)  # Time axis
y = np.zeros(len(t))  # Initialize signal

# Iterate through each time point
for j in range(len(t)):
    # Check if time falls within the current bit interval
    if t[j] >= a and t[j] <= a + 0.5:
        y[j] = n[i - 1]  # Assign voltage according to bit value
    # Check if time falls within the next bit interval
    elif t[j] >= a + 0.5 and t[j] <= i:
        pass  # Maintain the previous voltage level
    else:
        a += 1  # Move to the next bit interval
        i += 1  # Move to the next bit

# Plot the waveform
plot.plot(t, y, linewidth="2", color='red')
plot.grid(True)
plot.axis([0, N, -1.5, 1.5])
plot.title("Unipolar Return to Zero")
plot.show()
```
