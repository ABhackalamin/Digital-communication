```python
import numpy as np
import matplotlib.pyplot as plt

# Binary data used in MATLAB code
x = np.random.randint(2,size=10)
# Display binary data
print("Binary data:", x)
# Bits to polar
p = np.where(x == 0, -1, 1)
# Separation of even and odd sequences
even_seq = p[::2]

odd_seq = p[1::2]
# NRZ polar line coder signal generation
t = np.arange(0, len(x), 0.01)
even_ps = np.repeat(even_seq, len(t) // len(even_seq))
odd_ps = np.repeat(odd_seq, len(t) // len(odd_seq))

# Figure 1: NRZ Polar Line Coded Signal
plt.figure(1)
plt.subplot(211)
plt.plot(t, even_ps, 'r')
plt.title('Even Signal')
plt.subplot(212)
plt.plot(t, odd_ps, 'r')
plt.title('Odd Signal')
plt.tight_layout()

# Carrier signals generation
c1 = np.cos(2 * np.pi * 1 * t)
c2 = np.sin(2 * np.pi * 1 * t)

# Figure 2: Cosine and Sine Signal
plt.figure(2)
plt.subplot(211)
plt.plot(t, c1, 'r')
plt.title('Cosine Signal')
plt.subplot(212)
plt.plot(t, c2, 'b')
plt.title('Sine Signal')
plt.tight_layout()

# QPSK waveform generation
r1 = even_ps * c1
r2 = odd_ps * c2
qpsk_sig = r1 - r2

# Figure 3: QPSK Signal
plt.figure(3)
plt.subplot(311)
plt.plot(t, r1, 'r')
plt.title('Even Signal')
plt.subplot(312)
plt.plot(t, r2, 'b')
plt.title('Odd Signal')
plt.subplot(313)
plt.plot(t, qpsk_sig, 'b')
plt.title('QPSK Signal')
plt.tight_layout()

plt.show()
```
