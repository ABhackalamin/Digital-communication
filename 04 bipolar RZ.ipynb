```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the number of bits
N = 10

# Generate a random binary sequence of length N
n = np.random.randint(2, size=N)
print('The binary sequence is', n)

# Create an array to store the bipolar representation of the binary sequence
nn = np.zeros(N)

# Initialize a flag to alternate between positive and negative values
flag = True

# Convert the binary sequence to bipolar representation
for i in range(N):
    if n[i] == 1:  # If the current bit is 1
        if flag:  # If the flag is True
            nn[i] = 1  # Set the bipolar value to 1
            flag = False  # Set the flag to False for the next iteration
        else:  # If the flag is False
            nn[i] = -1  # Set the bipolar value to -1
            flag = True  # Set the flag to True for the next iteration
    else:  # If the current bit is 0
        nn[i] = 0  # Set the bipolar value to 0

print("The biplar code is : ",nn)        

# Initialize variables for waveform generation
i = 1  # Bit index
a = 0  # Time interval start
t = np.arange(0, N + 0.01, 0.01)  # Time axis with finer resolution
y = np.zeros(len(t))  # Initialize the waveform array

# Generate the Bipolar Return to Zero waveform
for j in range(len(t)):
    if t[j] >= a and t[j] <= a + 0.5:  # If the time falls within the first half of the bit interval
        y[j] = nn[i - 1]  # Assign the corresponding bipolar value
    elif t[j] >= a + 0.5 and t[j] <= i:  # If the time falls within the second half of the bit interval
        y[j] = 0  # Assign zero voltage
    else:  # If the time exceeds the current bit interval
        i += 1  # Move to the next bit
        a += 1  # Update the time interval start

# Plot the Bipolar Return to Zero waveform
plt.plot(t, y, linewidth=2, color='red')  # Plot the waveform
plt.grid(True)  # Add gridlines to the plot
plt.axis([0, N, -1.5, 1.5])  # Set the axis limits
plt.title('Bipolar Return to Zero')  # Set the title of the plot
plt.show()  # Display the plot
```
