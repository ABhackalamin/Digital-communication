```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square

fc = 20
fm = 2
fs = 1000
t = 1
n = np.arange(0, t, 1 / fs)
duty = 20

s = square(2 * np.pi * fc * n, duty / 100)
s[s < 0] = 0

plt.figure(1)

m = np.sin(2 * np.pi * fm * (n - 1))
period_samp = len(n) / fc
ind = np.arange(0, len(n), period_samp, dtype=int)
on_samp = int(np.ceil(period_samp * duty / 100))

pam = np.zeros_like(n)
for i in ind:
    pam[i:i + on_samp] = m[i]

plt.figure(1)
plt.subplot(1, 1, 1)
plt.plot(n, s,color='red')
plt.ylim([-0.2, 1.2])
plt.title('Carrier pulses Train used for Modulation')
plt.grid(True)
plt.figure(2)
plt.subplot(1, 1, 1)
plt.plot(n, m,color='red')
plt.ylim([-1.2, 1.2])
plt.title('Analog signal for Modulating')
plt.grid(True)
plt.figure(3)
plt.subplot(1, 1, 1)
plt.plot(n, pam,color='red')
plt.ylim([-1.2, 1.2])
plt.title('Pulse Amplitude Modulated Output')
plt.grid(True)
plt.show()
```
